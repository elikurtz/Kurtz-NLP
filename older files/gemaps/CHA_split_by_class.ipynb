```python
from ChaFile import *
import os
from pydub import AudioSegment as AS
```

```python
dir_healthy_wav = input("enter the (full) path to your directory of healthy WAV files --> ")
dir_healthy_cha = input("enter the (full) path to your directory of healthy CHA files --> ")

dir_AD_wav = input("enter the (full) path to your directory of AD WAV files --> ")
dir_AD_cha = input("enter the (full) path to your directory of AD CHA files --> ")

dir_MCI_wav = input("enter the (full) path to your directory of MCI WAV files --> ")
dir_MCI_cha = input("enter the (full) path to your directory of MCI CHA files --> ")
```

```python
def checker(wav, cha):
    files_from_dir = os.listdir(wav)
    files_from_dir_2 = os.listdir(cha)
    if len(files_from_dir) != len(files_from_dir_2):
        quit("not equal")


checker(dir_healthy_wav, dir_healthy_cha)
checker(dir_AD_wav, dir_AD_cha)
checker(dir_MCI_wav, dir_MCI_cha)
```

```python
import os
def exporter(wav, cha_dir, outpath):
    files_from_dir = os.listdir(cha_dir)
    files_from_dir_2 = os.listdir(wav)
    print(outpath)
    count = 0
    for q in files_from_dir:
        os.chdir(cha_dir)
        cha = ChaFile(q)
        alexa_times = []
        lines = cha.getLines()
        for i in lines:
            timestamp = i["bullet"] 
            alexa_times.append(timestamp)
        command_num = 1
        for t_slice in alexa_times:
            print(t_slice)
            os.chdir(wav)
            part = AS.from_file(files_from_dir_2[count])
            clip = part[t_slice[0]:t_slice[1]]
            name = (wav + "_participant" + files_from_dir_2[count][:-4] + "_command#" + str(command_num) + ".wav")
            print(command_num, name, count)
            if outpath == 1:
                os.chdir("C:\\Users\\EliKurtz\\VSCode_files\\NLP\\Kurtz-NLP\\healthy_clipped_commands")
                clip.export(name, format="wav")
            if outpath == 2:
                os.chdir("C:\\Users\\EliKurtz\\VSCode_files\\NLP\\Kurtz-NLP\\ad_clipped_commands")
                clip.export(name, format="wav")
                retval = os.getcwd()
                print(retval)
            if outpath == 3:
                os.chdir("C:\\Users\\EliKurtz\\VSCode_files\\NLP\\Kurtz-NLP\\mci_clipped_commands")
                clip.export(name, format="wav")
                retval = os.getcwd()
                print(retval)
            command_num += 1
        count += 1
```

```python
exporter(dir_healthy_wav, dir_healthy_cha, 1)
exporter(dir_AD_wav, dir_AD_cha, 2)
exporter(dir_MCI_wav, dir_MCI_cha, 3)
```
