```python
from ChaFile import *
import os
import pandas as pd
from pydub import AudioSegment as AS
```

```python
dir_1 = input("enter the (full) path to your directory of CHA files --> ")
dir_2 = input("enter the (full) path to the directory of audio files you wish to split using the CHA files you previously gave --> ")
files_from_dir = os.listdir(dir_1)
files_from_dir_2 = os.listdir(dir_2)
if len(files_from_dir) != len(files_from_dir_2):
    quit("not equal")
```

```python
count = 0
for q in files_from_dir:
    os.chdir(dir_1)
    cha = ChaFile(q)
    alexa_times = []
    lines = cha.getLines()
    for i in lines:
        timestamp = i["bullet"] 
        alexa_times.append(timestamp)
    command_num = 1
    for t_slice in alexa_times:
        print(t_slice)
        os.chdir(dir_2)
        part = AS.from_file(files_from_dir_2[count])
        clip = part[t_slice[0]:t_slice[1]]
        name = ("participant" + files_from_dir_2[count][:-4] + "_command#" + str(command_num) + ".wav")
        os.chdir("C:\\Users\\EliKurtz\\Desktop\\AD_Audio\\clipped_commands")
        clip.export(name, format="wav")
        print(command_num, name)
        command_num += 1
    count += 1
    
    # alexa_df = pd.DataFrame(alexa_times)
    # filename = (q + "_alexa_timestamps")
    # alexa_df.to_csv(r"C:\Users\EliKurtz\Desktop\AD_Audio\csv_timestamps" + "\\" + filename)  
 
```
